```python
import numpy as np
```

```python
lst_marks=[45, 67, 89, 90, 23, 56, 78, 99, 100, 34, 23, 45, 67, 88, 92, 54, 76, 84, 91, 87]
minimum,Q1,median,Q3,max=np.quantile(lst_marks, [0,0.25,0.5,0.75,1])
```

```python
minimum,Q1,median,Q3,max
```

```python
IQR=Q3-Q1
print("Interquartile range:", IQR)
```

```python
lower_fence=Q1 - 1.5 * IQR
upper_fence=Q3 + 1.5 * IQR
```

```python
lower_fence, upper_fence
```
