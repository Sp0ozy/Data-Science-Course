```python
import pandas as pd
df=pd.DataFrame({'color':['red','blue','green','green','red','blue']})
```

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder=LabelEncoder()
```

```python
lbl_encoder.fit_transform(df[['color']])
```

```python
lbl_encoder.transform([['red']])
```

```python
df = pd.DataFrame({'size': ['small', 'medium', 'large', 'medium', 'small', 'large']})
df
```

```python
# create instance of OrdinalEncoder and then perform fit.transform

encoder=OrdinalEncoder(categories=[['small','medium','large']])
```

```python
encoder.fit_transform(df[['size']])
```

```python
encoder.transform([['small']])
```

```python
df = pd.DataFrame({
    'city': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix','Los Angeles'],
    'price': [850000, 650000, 450000, 300000, 400000, 700000]
    })
```

```python
mean_price=df.groupby('city')['price'].mean()
mean_price
```

```python
df['city_encoded']=df['city'].map(mean_price)
df
```

```python
df[['price','city_encoded']]
```

```python
import seaborn as sns
df=sns.load_dataset('tips')
df
```

```python
mean_price=df.groupby('time')['total_bill'].mean()
mean_price
```

```python
df['time_encoded']=df['time'].map(mean_price)
```
